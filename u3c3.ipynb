        <h1 style="color: #fc5100; margin-top: 100px;">Motores de recomendación de filtrado colaborativo basados en redes neuronales</h1>
        <p style="font-size: 18px; color: #002060;">
            Hasta el momento solo hemos visto algoritmos clásicos de motores de recomendación. Ahora, vamos a pasar a ser adultos:
        </p>
        <div style="position:relative; width: 428px; margin: 50px auto; padding: 15px auto;">
            <figure style="width: 478px; height: 400px; padding: 0 40px; margin: 0; text-align: center; color: #002060;">
                <img style="width: 470px; height: 400px; padding: 0; margin: 0;" src="./recursos/imagenes/toy_story.gif">
            </figure>
        </div>
        <p style="font-size: 18px; color: #002060;">
            Hay dos maneras de aplicar <em>deep learning</em> a un sistema de recomendación.
        </p>
        <ol style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>Extracción de variables</strong>. Consiste en extraer información gracias a técnicas con redes neuronales. Para el ejemplo de H&M se ve claro. Podríamos usar <em>computer vision</em> para extraer características de la imágenes y poder usarlas como variables de contenido, o contextuales en un algoritmo clásico. O, podríamos usar técnicas de procesamiento de lenguaje natural para extraer información de la columna de descripción del producto.
            </li>
            <li style="margin: 20px 0;"><strong>Algoritmo de recomendación</strong>. En este caso, consiste en usar un modelo de redes neuronales, que de un valor sobre un producto con el cual luego pueda ordenar.
            </li>
        </ol>
        <p style="font-size: 18px; color: #002060;">
            Nos vamos a centrar en este taller en esta segunda opción. Gracias a <a href="https://arxiv.org/pdf/1708.05031.pdf" target="_blank">Neural Collaborative Filtering</a> tenemos un marco del que poder partir para esto.
        </p>
        <div style="position:relative; width: 458px; margin: 50px auto; padding: 15px auto;">
            <figure style="width: 508px; height: 400px; padding: 0 40px; margin: 0; text-align: center; color: #002060;">
                <img style="width: 500px; height: 400px; padding: 0; margin: 0;" src="./recursos/imagenes/neu_cf.png">
                <figcaption>Algoritmo de <em>Neural Collaborative Filtering</em>.</figcaption>
            </figure>
        </div>
        <p style="font-size: 18px; color: #002060;">
            La idea obtener un espacio latente de los usuarios y los productos a través de Embeddings y con la multiplicación elemento a elemento o la concatenación de estos, permitir que el algoritmo no solo aprenda las posibles relaciones lineales que puede haber entre los usuarios y los productos, si no que aprenda también si hay relaciones no lineales.
        </p>
        <p style="font-size: 18px; color: #002060;">
            Si lo que usamos es la multiplicación elemento a elemento de los vectores latentes, inherentemente estaremos "replicando" la factorización matricial. Es por eso que a este algoritmo se le denomina GMF de <em>Generalized Matrix Factorization</em>. Esta multiplicación la pasaremos finalmente por una neurona con una función de activación sigmoidal con el fin de que nos ofrezca una probibilidad sobre cada producto.
        </p>
        <p style="font-size: 18px; color: #002060;">
            Si lo que usamos es la concatenación de los vectores latentes, para poder extraer relación entre usuarios y productos (es decir, beneficiarnos de esa colaboración), tendremos que pasar ese vector por un MLP para que aprenda esta interacción. En este caso, y por la misma razón, pasaremos finalmente por una neurona.
        </p>
        <p style="font-size: 18px; color: #002060;">
            Finalmente, podemos unificar fuerzas en el <em>NeuMF</em> o <em>Neural Matrix Factorization</em>. Que es un <em>ensemble</em> de ambos modelos a los que finalmente se concatenan los resultados y pasa por una neurona.
        </p>
        <div style="position:relative; width: 458px; margin: 50px auto; padding: 15px auto;">
            <figure style="width: 508px; height: 400px; padding: 0 40px; margin: 0; text-align: center; color: #002060;">
                <img style="width: 500px; height: 400px; padding: 0; margin: 0;" src="./recursos/imagenes/neu_mf.png">
                <figcaption>Algoritmo de <em>Neural Matrix Factorization</em>.</figcaption>
            </figure>
        </div>
        <p style="font-size: 18px; color: #002060;">
            Para el ejemplo que vamos a tratar vamos a usar variables implícitas, es decir, 1s y 0s. Es por eso que el significado de lo que vamos a predecir es cuánto un producto es relevante para un usuario. Por ello además de entrenar el modelo con productos relevantes (por ejemplo todo el historial de compras de la persona), necesitamos entrenar el modelo con productos que no sean relevantes (es decir, que no haya comprado). En el paper nos indica que para coger estos valores se tomen X productos no comprados de manera aleatoria, en este taller vamos a ir un poco más allá. Con el conocimiento aprendido anteriormente vamos a calcular 3 productos más alejados para cada producto, y hacer que estos actúen como 0. Para ello vamos a extraer la lejanía de los productos a través de la factorización SVD y usar la des-similaridad del coseno.
        </p>

```python
from tensorflow.keras.layers import Concatenate, Input, Embedding, Flatten, Dense, Multiply
from tensorflow.keras import Model
from tensorflow.keras.optimizers import Adam, SGD
from typing import List
from sklearn.metrics import pairwise_distances
from sklearn.decomposition import TruncatedSVD
from tqdm import tqdm

import pandas as pd
import numpy as np
import tools
```

```python
# Leemos los datos
transactions_train = pd.read_csv(
    "recursos/datos/transactions_small_train.csv", 
    dtype={"article_id": str, "cid": str}
)
transactions_test_small = pd.read_csv(
    "recursos/datos/transactions_small_test.csv",
    dtype={"article_id": str, "cid": str}
)
```

```python
# Vamos a transformar los datos de entrenamiento en su forma 
# de matriz de interacción.
user_item_matrix = pd.pivot_table(
    data=transactions_train, 
    columns="article_id", 
    index="cid", 
    aggfunc=lambda _: 1,  # <-- Esto se añade para que en las celdas aparezca un 1.
    fill_value=0
)
user_item_matrix = user_item_matrix.astype(float)
user_item_matrix.head(3)
```

```python
# En este caso, dado que se tiene la interacción usuario-item, se podría utilizar
# la descripción de cada item a través de dicha matriz. Sin embargo, vamos a describirlo
# como en las otras unidades (a través de sus variables de contenido) para el cálculo de
# la diversity y poder comparar.
configuration = tools.read_config_data()

dtypes_articles = configuration.get("DTYPES_ARTICLES")
dtypes_articles = {
    key: eval(value) if value.startswith("np") else value for key, value in dtypes_articles.items()
}

path_articles = "./recursos/datos/articles.csv"
articles = pd.read_csv(path_articles, dtype=dtypes_articles)

content_columns_all = [
    "product_type_no", "graphical_appearance_no", "colour_group_code", "index_group_no", "index_code", "section_no", "garment_group_no"
]
content_columns = [
    content_columns_all[0], content_columns_all[2], 
    content_columns_all[4], content_columns_all[6]
]

articles_content = articles[["article_id", *content_columns]]
for cc in content_columns:
    articles_content[cc] = articles_content[cc].astype(str)

articles_content_ohe = pd.get_dummies(articles_content, columns=content_columns)
articles_content_ohe = articles_content_ohe.set_index("article_id")
articles_content_ohe.head(2)
```

```python
# Este es el mismo código de la unidad anterior.

def get_top_n_similarities(latent_description: pd.DataFrame, latent_id: str, n: int = 50) -> pd.DataFrame:
    """Get the top N anti-similar users for a given user.
    
    Apply the anti cosine similarity to calc anti-similarities.
    
    :param latent_description: Latent matrix description.
    :param latent_id: Given latent id to calc the similarities.
    :param n: Top n users to use as similars. By default is 50.
    :return: Dataframe with the anti-top n users or items similar to latent_id using
        cosine metrics.
    """
    
    # La funcion pairwise_distances permite calcular diferentes distancias
    # en este caso la distancia del coseno. 
    latent_id_arr = latent_description.loc[latent_id].values.reshape(1, -1)
    cosine_sims = pairwise_distances(
        X=latent_id_arr, 
        Y=latent_description.values,
        metric="cosine"
    )
    
    # Transforma en dataframe los resultados. Y elimina la similaridad con
    # su propio usuario.
    cosine_sims_df = pd.DataFrame(
        cosine_sims, 
        index=[latent_id], 
        columns=latent_description.index
    )
    cosine_sims_df = cosine_sims_df.drop(columns=[latent_id])
    
    # Para usar el método pandas.nlargest es necesario que el formato del dataframe
    # sea column-wise y no row-wise. Por eso hay que calcularlo sobre la traspuesta.
    top_n_cosine_sims_df = cosine_sims_df.T.nlargest(n=n, columns=latent_id)
    top_n_cosine_sims_df = top_n_cosine_sims_df.rename(columns={latent_id: f"mult_{latent_id}"})
    return top_n_cosine_sims_df
```

```python
# Instanciamos un SVD
n_latent = 150
svd = TruncatedSVD(n_components=n_latent, algorithm="arpack", tol=0)
svd.fit(user_item_matrix)
```

```python
# Extraemos la matriz de productos en el espacio latente.
v = svd.components_
v.shape
```

```python
# Transformamos en dataframe
df_v = pd.DataFrame(
    data=v, 
    index=[f"k{_}" for _ in range(n_latent)], 
    columns=user_item_matrix.columns
)
```

```python
# Obtenemos el top 3 productos más alejados entre si
dfs = []
for aid in tqdm(transactions_train["article_id"].unique()):
    top_sims = get_top_n_similarities(latent_id=aid, latent_description=df_v.T, n=3)  # Cambiar solo esto
    top_sims = top_sims.reset_index(drop=False)
    top_sims["aid"] = aid
    top_sims = top_sims[["aid", "article_id"]]
    dfs.append(top_sims)
distanced_articles = pd.concat(dfs)
```

```python
# Creamos el set de testeo añadiendo 50 veces más productos no comprados que
# los comprados para luego hacer el ranking.
df_articles = transactions_train[["article_id"]].drop_duplicates()
iterable = [
    (50, transactions_test_small, []),
]

for neg_rate, df, dfs_list in iterable:
    for name, group in df.groupby("cid"):
        # Filtra por los productos que no han sido comprados.
        temp = pd.merge(
            left=df_articles, right=group,
            on="article_id", how="left", indicator=True
        )
        temp = temp[temp["_merge"] == "left_only"]
        
        # Obten de los productos no comprados una muestra
        # que se usará como muestra de productos negativos.
        n_items = group["article_id"].nunique()
        temp_neg = temp[["article_id"]].sample(
            n=n_items*neg_rate
        )
        
        # Formatea el resultado
        temp_neg["cid"] = name
        temp_neg["interested"] = 0
        temp_neg = temp_neg[["cid", "article_id", "interested"]]
        group["interested"] = 1
        dfs_list.extend([group, temp_neg])
```

```python
# Crea los dataframes de entrenamiento y test.
distanced_articles = distanced_articles.rename(columns={"article_id": "distanced_aid"})
train = pd.merge(
    left=transactions_train, right=distanced_articles, 
    left_on="article_id", right_on="aid", 
    how="inner"
)
train_01 = train[["cid", "article_id"]].drop_duplicates()
train_01["interested"] = 1
train_02 = train[["cid", "distanced_aid"]].drop_duplicates()
train_02 = train_02.rename(columns={"distanced_aid": "article_id"})
train_02["interested"] = 0
train = pd.concat([train_01, train_02])
test = pd.concat(iterable[-1][-1])

# Crea un identificador para cada usuario y para cada item.
train["#cid"] = train.groupby("cid").ngroup()
train["#aid"] = train.groupby("article_id").ngroup()

# Añade el identificador al set de test.
test = pd.merge(
    left=test, right=train[["#cid", "cid"]].drop_duplicates(), 
    how="left", on=["cid"]
)
test = pd.merge(
    left=test, right=train[["#aid", "article_id"]].drop_duplicates(), 
    how="left", on=["article_id"]
)
```

```python
train.sample(n=5)
```

```python
# Parametros de entrada a los modelos
train = train.sort_values(
    by=["cid", "interested"], 
    ascending=[False, False]
).drop_duplicates(
    subset=["cid", "article_id"], 
    keep="first"
)
m_users = train["#cid"].nunique()
n_items = train["#aid"].nunique()
gmf_latent_dim = 10
user_latent_dim = 5
item_latent_dim = 15
users = train["#cid"].values
items = train["#aid"].values
labels = train["interested"].values
users_test = test["#cid"].values
items_test = test["#aid"].values
learning_rate = 1e-3
```

<h2 style="color: #f47b4c; margin-top: 100px;">GMF</h2>

```python
def ncf_gmf(
    m_users: int, 
    n_items: int,
    latent_dim: int,
    user_input,
    item_input,
):
    name = "NeuCF-GMF"
    user_lyr_emb = Embedding(
        name="user_lyr_emb_ncf_gmf", input_dim=m_users, output_dim=latent_dim,
        embeddings_initializer="normal", embeddings_regularizer=None
    )
    item_lyr_emb = Embedding(
        name="item_lyr_emb_ncf_gmf", input_dim=n_items, output_dim=latent_dim,
        embeddings_initializer="normal", embeddings_regularizer=None
    )

    user_emb = user_lyr_emb(user_input)
    item_emb = item_lyr_emb(item_input)
    user_latent_flat = Flatten(name="user_latent_flat_ncf_gmf")(user_emb)
    item_latent_flat = Flatten(name="item_latent_flat_ncf_gmf")(item_emb)

    ui_score_vector = Multiply(name="ui_ew_prod")([user_latent_flat, item_latent_flat])
    
    return name, ui_score_vector
```

```python
user_input = Input(name="user_input", shape=(1,), dtype="int32")
item_input = Input(name="item_input", shape=(1,), dtype="int32")
ncf_gmf_name, ncf_gmf_last_layer = ncf_gmf(
    m_users=m_users,
    n_items=n_items,
    latent_dim=gmf_latent_dim,
    user_input=user_input,
    item_input=item_input,
)
ncf_gmf_score = Dense(
    name="ncf_gmf_score", activation="sigmoid", 
    units=1, kernel_initializer="lecun_uniform"
)(ncf_gmf_last_layer)

ncf_gmf_model = Model(name=ncf_gmf_name, inputs=[user_input, item_input], outputs=ncf_gmf_score)
ncf_gmf_model.compile(
    optimizer=Adam(learning_rate=learning_rate),
    loss="binary_crossentropy", 
    metrics=["accuracy"],
)
ncf_gmf_model.summary()
```

```python
ncf_gmf_model.fit(
    x=[users, items], y=labels, 
    batch_size=1032, epochs=10, shuffle=True
)
gmf_model_weights = f"./recursos/pesos/gmf.h5"
ncf_gmf_model.save_weights(gmf_model_weights, overwrite=True)
```

```python
test["score"] = ncf_gmf_model.predict(x=[test["#cid"].values, test["#aid"].values])
```

```python
# Eliminamos los artículos que ya ha comprado el usuario.
recomendations_df = pd.merge(
    left=test, right=train[["cid", "article_id"]], 
    on=["cid", "article_id"], how="left", indicator=True
)
recomendations_df = recomendations_df[recomendations_df["_merge"] == "left_only"]
recomendations_df = recomendations_df.sort_values(by=["cid", "score"], ascending=[False, False])
recomendations_df = recomendations_df.groupby("cid").head(5)
recomendations_df = recomendations_df[["cid", "article_id", "score"]]
```

```python
tools.get_metrics(
    name=ncf_gmf_name, 
    recommendations=recomendations_df, 
    sort_column="score",
    true_labels=test[["cid", "article_id", "interested"]], 
    articles_df=articles_content_ohe
)
```

<h2 style="color: #f47b4c; margin-top: 100px;">MLP</h2>

```python
def ncf_mlp(
    m_users: int, 
    n_items: int,
    user_latent_dim: int,
    item_latent_dim: int,
    user_input,
    item_input,
):
    name = "NeuCF-MLP"
    user_lyr_emb = Embedding(
        name="user_lyr_emb_ncf_mlp", input_dim=m_users, output_dim=user_latent_dim,
        embeddings_initializer="normal", embeddings_regularizer=None
    )
    item_lyr_emb = Embedding(
        name="item_lyr_emb_ncf_mlp", input_dim=n_items, output_dim=item_latent_dim,
        embeddings_initializer="normal", embeddings_regularizer=None
    )

    user_emb = user_lyr_emb(user_input)
    item_emb = item_lyr_emb(item_input)
    user_latent_flat = Flatten(name="user_latent_flat_ncf_mlp")(user_emb)
    item_latent_flat = Flatten(name="item_latent_flat_ncf_mlp")(item_emb)

    ui_concat = Concatenate(name="ui_concat")([user_latent_flat, item_latent_flat])
    
    mlp_01 = Dense(
        name="mlp_01", activation="relu", 
        units=64, kernel_initializer="glorot_uniform"
    )
    ui_mlp_01 = mlp_01(ui_concat)
    mlp_02 = Dense(
        name="mlp_02", activation="relu", 
        units=32, kernel_initializer="glorot_uniform"
    )
    ui_mlp_02 = mlp_02(ui_mlp_01)
    mlp_03 = Dense(
        name="mlp_03", activation="relu", 
        units=16, kernel_initializer="glorot_uniform"
    )
    ui_mlp_03 = mlp_03(ui_mlp_02)
    mlp_04 = Dense(
        name="mlp_04", activation="relu", 
        units=8, kernel_initializer="glorot_uniform"
    )
    ui_mlp_04 = mlp_04(ui_mlp_03)
    
    return name, ui_mlp_04
```

```python
user_input = Input(name="user_input", shape=(1,), dtype="int32")
item_input = Input(name="item_input", shape=(1,), dtype="int32")
ncf_mlp_name, ncf_mlp_last_layer = ncf_mlp(
    m_users=m_users,
    n_items=n_items,
    user_latent_dim=user_latent_dim,
    item_latent_dim=item_latent_dim,
    user_input=user_input,
    item_input=item_input,
)
ncf_mlp_score = Dense(
    name="ncf_mlp_score", activation="sigmoid", 
    units=1, kernel_initializer="lecun_uniform"
)(ncf_mlp_last_layer)

ncf_mlp_model = Model(name=ncf_mlp_name, inputs=[user_input, item_input], outputs=ncf_mlp_score)
ncf_mlp_model.compile(
    optimizer=Adam(learning_rate=learning_rate),
    loss="binary_crossentropy", 
    metrics=["accuracy"],
)
ncf_mlp_model.summary()
```

```python
ncf_mlp_model.fit(
    x=[users, items],
    y=labels, 
    batch_size=1032, epochs=10, shuffle=True
)
mlp_model_weights = f"./recursos/pesos/mlp.h5"
ncf_mlp_model.save_weights(mlp_model_weights, overwrite=True)
```

```python
test["score"] = ncf_mlp_model.predict(x=[users_test, items_test])
```

```python
recomendations_df = pd.merge(
    left=test, right=train[["cid", "article_id"]], 
    on=["cid", "article_id"], how="left", indicator=True
)
recomendations_df = recomendations_df[recomendations_df["_merge"] == "left_only"]
recomendations_df = recomendations_df.sort_values(by=["cid", "score"], ascending=[False, False])
recomendations_df = recomendations_df.groupby("cid").head(5)
recomendations_df = recomendations_df[["cid", "article_id", "score"]]
```

<h2 style="color: #f47b4c; margin-top: 100px;">NeuMF</h2>

```python
user_input = Input(name="user_input", shape=(1,), dtype="int32")
item_input = Input(name="item_input", shape=(1,), dtype="int32")
_, ncf_mlp_last_layer = ncf_mlp(
    m_users=m_users,
    n_items=n_items,
    user_latent_dim=user_latent_dim,
    item_latent_dim=item_latent_dim,
    user_input=user_input,
    item_input=item_input,
)
_, ncf_gmf_last_layer = ncf_gmf(
    m_users=m_users,
    n_items=n_items,
    latent_dim=gmf_latent_dim,
    user_input=user_input,
    item_input=item_input,
)
ensemble_layer = Concatenate(name="ensemble_layer")([ncf_gmf_last_layer, ncf_mlp_last_layer])
ncf_neumf_score = Dense(
    name="ncf_neumf_score", activation="sigmoid", 
    units=1, kernel_initializer="lecun_uniform"
)(ensemble_layer)

ncf_neumf_model = Model(name="NCF-NeuMF", inputs=[user_input, item_input], outputs=ncf_neumf_score)

ncf_gmf_model.load_weights(gmf_model_weights)
ncf_mlp_model.load_weights(mlp_model_weights)
ncf_neumf_model.get_layer("user_lyr_emb_ncf_gmf").set_weights(
    ncf_gmf_model.get_layer("user_lyr_emb_ncf_gmf").get_weights()
)
ncf_neumf_model.get_layer("item_lyr_emb_ncf_gmf").set_weights(
    ncf_gmf_model.get_layer("item_lyr_emb_ncf_gmf").get_weights()
)
ncf_neumf_model.get_layer("user_lyr_emb_ncf_mlp").set_weights(
    ncf_mlp_model.get_layer("user_lyr_emb_ncf_mlp").get_weights()
)
ncf_neumf_model.get_layer("item_lyr_emb_ncf_mlp").set_weights(
    ncf_mlp_model.get_layer("item_lyr_emb_ncf_mlp").get_weights()
)
ncf_neumf_model.get_layer("mlp_01").set_weights(
    ncf_mlp_model.get_layer("mlp_01").get_weights()
)
ncf_neumf_model.get_layer("mlp_02").set_weights(
    ncf_mlp_model.get_layer("mlp_02").get_weights()
)
ncf_neumf_model.get_layer("mlp_03").set_weights(
    ncf_mlp_model.get_layer("mlp_03").get_weights()
)
ncf_neumf_model.get_layer("mlp_04").set_weights(
    ncf_mlp_model.get_layer("mlp_04").get_weights()
)

ncf_gmf_last_layer_weights = ncf_gmf_model.get_layer("ncf_gmf_score").get_weights()
ncf_mlp_last_layer_weights = ncf_mlp_model.get_layer("ncf_mlp_score").get_weights()
layer_weigths = .5 * np.concatenate((ncf_gmf_last_layer_weights[0], ncf_mlp_last_layer_weights[0]), axis=0)
neuron_weights = .5 * (ncf_gmf_last_layer_weights[1] + ncf_mlp_last_layer_weights[1])
ncf_neumf_model.get_layer("ncf_neumf_score").set_weights(
    [layer_weigths, neuron_weights]
)
ncf_neumf_model.compile(
    optimizer=SGD(learning_rate=learning_rate),
    loss="binary_crossentropy", 
    metrics=["accuracy"],
)
ncf_neumf_model.summary()
```
