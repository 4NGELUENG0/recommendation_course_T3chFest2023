        <h1 style="color: #fc5100; margin-top: 100px;">Motores de recomendación de filtrado colaborativo basados en memoria</h1>
        <p style="font-size: 18px; color: #002060;">
            Los métodos de filtrado colaborativo basados en memoria utilizan la <strong>matriz completa de iteracción de usuario-ítem</strong> para generar recomendaciones.
        </p>
        <p style="font-size: 18px; color: #002060;">
        Se pueden clasificar en dos grupos:  
        <ul style="text-align: left; font-size: 18px; color: #002060;">
                <li style="margin: 20px 0;"><strong>Centrado en el usuario</strong> o <em>user-based CF</em>: Miden la similitud entre el usuario objetivo y otros usuarios para rellenar la matriz de interacción.</li>
                <li style="margin: 20px 0;"><strong>Centrado en los artículos</strong> o <em>item-based CF</em>: Miden la similitud entre los ítems con los que el usuario iteractúan y otros ítems para rellenar la matriz de interacción.</li>
        </ul>
        <p style="font-size: 18px; color: #002060;">
            Antes de hablar de estas dos maneras de afrontar el problema de recomendar, vamos a hablar de distancias y de formas de recomendar. La idea detrás de estos algoritmos, de este capítulo, es rellenar los huecos de la matriz de interacción de <em>user-item</em> a través de usuarios parecidos o productos parecidos. Estos productos, o usuarios, parecidos se calculan en base a distancias, es decir, un usuario se parece más a mi a medida los ratings que le damos a los productos sean similares. O, dos productos se parecen más si todos los usuarios que añaden un rating aportan ratings similares. Hay muchas formas de calcular distancias, en este taller vamos a ver dos.
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>Similaridad del coseno</strong>: De esta ya hemos hablado en las métricas. Se trata de, geométricamente, ver si dos vectores son paralelos entre sí. Cuanto más paralelos son, es decir, cuanto la similaridad del coseno más sea 1, más se pareceran entre sí.</li>
            <li style="margin: 20px 0;"><strong>Coeficiente de Jaccard</strong>: Esta mide la similitud entre dos conjuntos basándose en el número de elementos que tienen en común dividido por el número total de elementos que son únicos para cada conjunto. Esta es conveniente usarla cuando no se tienen ratings, o lo que es lo mismo, cuando en la matriz <em>user-item</em> solo se tienen 0 y 1.</li>
        </ul>
        <p style="font-size: 18px; color: #002060;">
            Cuando hablamos de formas de recomendar nos referimos a la manera en la que se rellena esa matriz de interacción. La forma más básica, o <em>naive</em>, en la que no se tienen en cuenta los usuarios o items más parecidos, sería poner la media de rating del artículo en un hueco, o la media de rating de la persona. Esta es la técnica más sencilla, pero no explota bien la importancia de los usuarios parecidos.<br>
            Otra manera de recomendar, y esta si tiene en cuenta la similitud, es utilizar el valor de la similitud (que está acotado entre 0 y 1), multiplicarlo por el rating que cada persona ha dado al artículo y dividirlo luego entre la cantidad de personas que ha dado rating al artículo. Esto va a provocar que los ratings de las personas que más se parezcan entre sí impacten más en la predicción del rating que una persona le daría a ese artículo. Con esto luego se pueden ordenar los artículos predichos con mayor valor y recomendar el top K artículos. En el caso de que se tenga en la matriz de interacción solo 0 y 1, luego se puede establecer un límite, o <em>threshold</em> sobre el cual se considera que un item va a ser un 1.<br>
            A esta forma que se ha comentado antes, también se le puede añadir una previa selección de el top Y usuarios o items más cercanos, sobre los cuales calcular el rating de la manera explicada.<br>
            Dos formas más de recomendar quedan, una es la siguiente. Tras haber seleccionado el top Y de usuarios o items, y sin tener en cuenta el coeficiente de semejanza, calcular la media del artículo. Y la última, también tras haber seleccionado el top Y, en este caso de usuarios, contar los items que compran estos usuarios y recomendar aquellos que más se repitan. La idea es que sea un recomendador basado en popularidad, pero sobre lo que es popular en un grupo de personas que se parece a ti.
        </p>
        <p style="font-size: 18px; color: #002060;">
            En este taller vamos a ver la forma en la que se toman los Y usuarios más cercanos y luego se pondera el resultado. Además, en la práctica un recomendador <em>user-based</em> e <em>item-based</em> se solucionan igual, es por eso que solo vamos a ver el <em>user-based</em> a nivel técnico. Y luego se observarán las métricas sobre un <em>item-based</em> previamente ejecutado, una de las razones de dejar este previamente ejecutado es además que es menos costoso computacionalmente, por si el alumno decide ejecutarlo durante el taller.
        </p>

Es decir, para mi un 5/5 puede ser muy dificil darlo y mi máxima puntuación es un 4, sin embargo, para otra persona dar 5/5 es lo más normal del mundo. Para ello se recurre a eliminar este sesgo, o <em>bias</em> en los datos.

```python
import pandas as pd
import numpy as np
import tools

from sklearn.metrics.pairwise import pairwise_distances

# pd.set_option("display.max_columns", None)
```

```python
configuration = tools.read_config_data()

dtypes_articles = configuration.get("DTYPES_ARTICLES")
dtypes_articles = {
    key: eval(value) if value.startswith("np") else value for key, value in dtypes_articles.items()
}

path_articles = "./recursos/datos/articles.csv"
articles = pd.read_csv(path_articles, dtype=dtypes_articles)
content_columns_all = [
    "product_type_no", "graphical_appearance_no", "colour_group_code", "index_group_no", "index_code", "section_no", "garment_group_no"
]
content_columns = [
    content_columns_all[0], content_columns_all[1], 
    content_columns_all[2], content_columns_all[3]
]
articles_content = articles[["article_id", *content_columns]]
for cc in content_columns:
    articles_content[cc] = articles_content[cc].astype(str)
articles_content_ohe = pd.get_dummies(articles_content, columns=content_columns)
articles_content_ohe = articles_content_ohe.set_index("article_id")
```

```python
transactions_small_train = pd.read_csv(
    "./recursos/datos/transactions_small_train.csv", 
    dtype={
        "cid": str, 
        "article_id": str
    }
)
transactions_small_test = pd.read_csv(
    "./recursos/datos/transactions_small_test.csv", 
    dtype={
        "cid": str, 
        "article_id": str, 
        "interested": int
    }
)
```

```python
# Vamos a transformar los datos de entrenamiento en su forma 
# de matriz de interacción.
user_item_matrix = pd.pivot_table(
    data=transactions_small_train, 
    columns="article_id", 
    index="cid", 
    aggfunc=lambda _: 1,  # <-- Esto se añade para que en las celdas aparezca un 1.
    fill_value=0
)
user_item_matrix = user_item_matrix.astype("int8")
user_item_matrix.head(3)
```

```python
# STUDENT-CODE

def get_top_n_similarities(user_item: pd.DataFrame, cid: str, n: int = 50) -> pd.DataFrame:
    """Get the top N similar users for a given user.
    
    Apply the jaccard similarity to calc similarities.
    
    :param user_item: Interaction matrix.
    :param cid: Given user to calc the similarities.
    :param n: Top n users to use as similars. By default is 50.
    :return: Dataframe with the top n users similar to cid using
        jaccard metrics.
    """
    
    # La funcion pairwise_distances permite calcular diferentes distancias
    # en este caso la distancia de jaccard. Esta función sin embargo no devuelve
    # un resultado de similaridad, si no de des-similaridad, es por ello que
    # es necesario hacer 1 - jaccard.
    cid_arr = user_item.loc[cid].values.reshape(1, -1)
    jaccard_sims = 1 - pairwise_distances(
        X=cid_arr, 
        Y=user_item.values,
        metric="jaccard"
    )
    
    # Transforma en dataframe los resultados. Y elimina la similaridad con
    # su propio usuario.
    jaccard_sims_df = pd.DataFrame(
        jaccard_sims, 
        index=[cid], 
        columns=user_item.index
    )
    jaccard_sims_df = jaccard_sims_df.drop(columns=[cid])
    
    # Para usar el método pandas.nlargest es necesario que el formato del dataframe
    # sea column-wise y no row-wise. Por eso hay que calcularlo sobre la traspuesta.
    top_n_jaccard_sims_df = jaccard_sims_df.T.nlargest(n=n, columns=cid)
    top_n_jaccard_sims_df = top_n_jaccard_sims_df.rename(columns={cid: f"mult_{cid}"})
    return top_n_jaccard_sims_df
```

```python
# STUDENT-CODE

# Calcula las similaridades para un usuario específico
top_n_sims = get_top_n_similarities(user_item=user_item_matrix, cid="0000039")
top_n_sims.head(5)
```

```python
# STUDENT-CODE

def get_top_k_recommendations(
    user_item: pd.DataFrame,
    similarities: pd.DataFrame,
    cid: str,
    k: int = 5
) -> pd.DataFrame:
    """Get top k article recommendations for a given user.
    
    :param user_item: Interaction matrix.
    :param similarities: Dataframe with the top n users similar to cid using
        jaccard metrics.
    :param cid: Given user to calc the similarities.
    :param k: Top k recommendations for the user. By default is 5.
    :return: Dataframe with k recommendation for the user.
    """
    # Filtramos a través de una unión los usuarios similares.
    sim_users = pd.merge(
        left=user_item,
        right=similarities,
        left_index=True, right_index=True,
        how="inner"
    )
    
    # Multiplicamos por el coeficiente de similaridad todos los artículos
    # de los usuarios.
    sim_users = sim_users[user_item.columns].mul(
        sim_users[f"mult_{cid}"], axis=0
    )
    
    # Para calcular la media, primero obten la suma de los productos
    # y luego la suma de los usuarios que han comprado ese producto.
    # Es decir, la suma de aquellos usuarios cuyas celdas no son 0.
    articles_sum = sim_users.sum(axis=0)
    article_count = (sim_users != 0).sum(axis=0)
    article_mean = articles_sum.div(article_count, fill_value=0)
    article_mean = article_mean.to_frame(name=cid)
    
    # Antes de obtener el top k de artículos a recomendar hay que eliminar
    # los productos que el usuario ha comprado ya.
    cid_articles = user_item.loc[cid]
    cid_articles = cid_articles[cid_articles != 0]
    article_mean = article_mean[~article_mean.index.isin(cid_articles.index)]
    
    # Finalmente obten el top k de artículos a recomendar
    return article_mean.nlargest(n=k, columns=cid)
```

```python
# STUDENT-CODE

# Calcula las recomendaciones para un usuario específico
recommendations = get_top_k_recommendations(
    user_item=user_item_matrix,
    cid="0000039",
    similarities=top_n_sims
)
recommendations
```

```python
# A partir de aquí se va a implementar un recomendador item-based
# para observar las métricas que se obtienen. Lo primero que se tiene
# que hacer, para su aliviar la carga cognitiva, es transponer la matriz
# user-item, para tener una item-user.
item_user_matrix = user_item_matrix.T

# Cálculo de las similaridades de Jaccard.
jaccard_sims = 1 - pairwise_distances(
    X=item_user_matrix.values, 
    metric="jaccard"
)
jaccard_sims_df = pd.DataFrame(
    jaccard_sims, 
    index=item_user_matrix.index, 
    columns=item_user_matrix.index
)

# Para vectorizar las operaciones se procede a despivotar la información.
jaccard_sims_df = jaccard_sims_df.unstack()
jaccard_sims_df.index.names = ["aid_01", "article_id"]
jaccard_sims_df = jaccard_sims_df.to_frame(name="jaccard_sim")
jaccard_sims_df = jaccard_sims_df.reset_index(drop=False)

# Ahora sí, se eliminan las similaridades entre el mismo par de artículos.
jaccard_sims_df = jaccard_sims_df[jaccard_sims_df["aid_01"] != jaccard_sims_df["article_id"]]

# Se obtiene el top 50 artículos más similar entre si.
jaccard_sims_df = jaccard_sims_df.sort_values(
    by=["aid_01", "jaccard_sim"], ascending=[False, False]
)
jaccard_sims_df = jaccard_sims_df.groupby("aid_01").head(50)

# Se une el vector asociado a cada item similar, con el objetivo de calcular
# el resto de "ratings".
item_user_sims = pd.merge(
    left=jaccard_sims_df, 
    right=item_user_matrix, how="inner", 
    left_on="article_id", right_index=True
)

# En este punto, vamos a filtrar por los usuarios que vamos a testear. Aprovechándonos que
# para un item-based el "rating" de un artículo a un usuario se da, en base a los productos
# similares, en función de los ratings que ese mismo usuario ha dado a otros artículos.
test_cids = transactions_small_test["cid"].unique().tolist()
item_user_sims = item_user_sims[["aid_01", "jaccard_sim", *test_cids]]

# Multiplicamos el valor de la similaridad para cada usuario.
item_user_sims_aux = item_user_sims[test_cids].mul(
    item_user_sims["jaccard_sim"], axis=0
)

# Transformamos el dataframe.
item_user_sims_aux["aid_01"] = item_user_sims["aid_01"]
item_user_sims_aux = item_user_sims_aux.set_index("aid_01")

# Calculamos la suma para cada artículo y el total de artículos similares
# que el usuario ha "rateado". Para luego calcular la media.
item_user_sims_sum = item_user_sims_aux.groupby(item_user_sims_aux.index).sum()
item_user_sims_count = item_user_sims_aux.astype(bool).groupby(item_user_sims_aux.index).sum()
article_recoms = item_user_sims_sum / item_user_sims_count

# Despivotamos la información de los ratings de los artículos
article_recoms = article_recoms.unstack()
article_recoms.index.names = ["cid", "article_id"]
article_recoms = article_recoms.to_frame(name="jaccard_sim")
article_recoms = article_recoms.reset_index(drop=False)

# Eliminamos la información de los productos que esos usuarios ya han comprado.
# Estoy se realiza a través de un anti left join.
article_recoms = pd.merge(
    left=article_recoms, right=transactions_small_train, 
    on=["cid", "article_id"], how="left", 
    indicator=True
)
article_recoms = article_recoms[article_recoms["_merge"] == "left_only"]

# Ordenamos por "cid" y por "jaccard_sim" de mayor a menor.
article_recoms = article_recoms.sort_values(
    by=["cid", "jaccard_sim"], ascending=[False, False]
)

# Obtenemos el top 5 de productos recomendados para cada usuario.
recommendations = article_recoms.groupby("cid").head(5)
recommendations.head(6)
```

```python
tools.get_metrics(
    name="CF-ItemBased", recommendations=recommendations, sort_column="jaccard_sim",
    true_labels=transactions_small_test, articles_df=articles_content_ohe
)
```

<p style="font-size: 18px; color: #002060;">
   Estos algoritmos tienen la ventaja de ser independientes de las características de los ítems y de los usuarios; podríamos implementarlos aun no contando con una base de datos con variables extra, sólo nos hacen falta las interacciones entre usuarios e ítems. Esta es, además, la primera familia de algoritmos que se aprovecha de lo que ha puntuado o lo que ha realizado el resto de personas. Sin embargo sufre de los siguiente problemas:
</p>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><em>Cold start</em>: Un nuevo cliente, o un nuevo artículo, que no tenga ninguna interacción previa no se puede ver benficiado de esta familia de algoritmos.</li>
    <li style="margin: 20px 0;">Escalabilidad: En este taller hemos reducido la cantidad de usuarios y la cantidad de productos sobre la que entrenamos y sobre la que recomendamos, esto hace accesible esta manera de recomendar. Sin embargo, a medida que aumenta el tamaño de la matriz el problema se hace cada vez menos escalable a una solución robusta.</li>
    <li style="margin: 20px 0;">Elección del número de usuarios o productos similares entre sí que optimize el resultado.</li>
    <li style="margin: 20px 0;">¿Qué hacemos con los usuarios que sólo han comprado / evaluado un elemento?</li>
</ul>
