        <h1 style="color: #fc5100; margin-top: 100px;">¿Qué tipos de motores de recomendación existen?</h1>
        <p style="font-size: 18px; color: #002060;">
            En esta sección se van a introducir las diferentes formas que hay de construir un motor de recomendación. A saber:
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>Métodos basados en contenido</strong> o <em>Content-based methods</em>.</li>
            <li style="margin: 20px 0;"><strong>Métodos de filtrado colaborativo</strong> o <em>Collaborative filtering methods</em>. Que a su vez se dividen en:</li>
            <ul style="text-align: left; font-size: 18px; color: #002060;">
                <li style="margin: 20px 0;"><strong>Basados en memoria</strong> o <em>Memory-based</em>.</li>
                <ul style="text-align: left; font-size: 18px; color: #002060;">
                    <li style="margin: 20px 0;"><strong>Centrado en el usuario</strong> o <em>user-based CF</em>.</li>
                    <li style="margin: 20px 0;"><strong>Centrado en los artículos</strong> o <em>item-based CF</em>.</li>
                </ul>
                <li style="margin: 20px 0;"><strong>Basados en modelos</strong> o <em>Model-based</em>.</li>
            </ul>
            <li style="margin: 20px 0;"><strong>Modelos híbridos</strong>.</li>
        </ul>
        <h2 style="color: #f47b4c; margin-top: 100px;">Métodos basados en contenido</h2>
        <p style="font-size: 18px; color: #002060;">
            Este tipo de modelos no tienen en cuenta la matriz de usuarios y productos. Para ello, esto modelos requieren de variables contextuales de usuarios o de los productos para poder hacer las recomendaciones. Estas pueden ser regresiones o clasificaciones (se puede obtener el rating que se le va a dar a un producto o si el producto puede gustar o no). Por ejemplo, se puede obtener un recomendador basado en el contenido de los libros o con los géneros que se le asocian a una película y la edad de las personas que ven esas películas.
        </p>
        <h2 style="color: #f47b4c; margin-top: 100px;">Métodos de filtrado colaborativo</h2>
        <p style="font-size: 18px; color: #002060;">
            Estos métodos se basan exclusivamente en las interacciones que hay entre los usuarios y los artículos. Es decir, que las interacciones que hayan tenido los usuarios con los productos es suficiente para poder realizar recomendaciones. Estas relaciones se guardan en lo que se conoce como la matriz de usuarios y productos o <em>user-item matrix</em>. Como se ha visto, estos se dividen en:
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><em>Memory-based</em>: Estos modelos solo tienen en cuenta las interacciones de la matriz de usuarios y productos.</li>
            <ul style="text-align: left; font-size: 18px; color: #002060;">
                <li style="margin: 20px 0;">Cuando se habla de <em>user-based CF</em> lo que quiere decir es que la <strong>similitud</strong> se realiza entre los usuarios. Esto quiere decir que se buscan usuarios que se parezcan entre sí para poder ofrecer una recomendación. Por ejemplo, usuarios que hayan comprado los mismos artículos, o, usuarios que hayan puntuado unos libros de manera similar.</li>
                <li style="margin: 20px 0;">Cuando se habla de <em>item-based CF</em> lo que quiere decir es que la <strong>similitud</strong> se realiza entre los artículos. Esto quiere decir que se buscan artículos que se parezcan entre sí para poder ofrecer una recomendación. Por ejemplo, dos libros se parecerán más si todas las personas le dan puntuaciones similares.</li>
            </ul>
            <li style="margin: 20px 0;"><em>Model-based</em>: Estos modelos también tienen en cuenta las interacciones de la matriz de usuarios y productos, solo que estos modelos se basan en que estas interacciones están soportadas por un modelo latente que define la matriz de interacción.</li>
        </ul>
        <h2 style="color: #f47b4c; margin-top: 100px;">Modelos híbridos</h2>
        <p style="font-size: 18px; color: #002060;">
            El tipo de modelo más usado. Mezcla de los otros métodos, usa información de la matriz de usuarios e items mientras a su vez añade información contextual. El motivo de que sea de los métodos más usados es porque con el se pueden ir construyendo modelos que vayan paliando las carencias de unos y otros. Carencias (o problemas asociados) que se verán en el siguiente capítulo.
        </p>
