        <h1 style="color: #fc5100; margin-top: 100px;">¿Qué métricas uso para un motor de recomendación?</h1>
        <p style="font-size: 18px; color: #002060;">
            Las métricas son esa conexión con el caso de uso, o con negocio, que resuelve el problema. Un problema de ML puede estar muy bien formulado, pero, si no se transmite con la métrica, o métricas, correcta la solución, esta puede perder toda su usabilidad. Dos ejemplos de la importancia de las métricas y su conexión con negocio.<br>
            <strong>Ejemplo 1:</strong> Imagina que el Codryceps se hace real:
        </p>
        <div style="position:relative; width: 458px; margin: 50px auto; padding: 15px auto;">
            <figure style="width: 458px; height: 450px; padding: 0 40px; margin: 0; text-align: center; color: #002060;">
                <img style="width: 450px; height: 450px; padding: 0; margin: 0;" src="./recursos/imagenes/tlou.gif">
            </figure>
        </div>
        <p style="font-size: 18px; color: #002060;">
            Eres la persona encargada de hacer un sistema de clasificación que detecte personas infectadas para dejarlas pasar, o no, a una zona de cuarentena. Si entran personas infectadas puede hacer que la cuarentena se caiga y esa zona segura se pierda.<br>
            Imagina que desarrollas un sistema de clasificación con una precisión de 99%, es decir, que cuando predices que una persona está infectada aciertas 99 veces de cada 100. ¡Nada mal eeh, artista!. Sin embargo, lo que esa métrica es incapaz de ver es ¿cuántas personas infectadas estoy dejando de predecir como infectadas? o de otro modo, ¿cuántas personas infectadas estoy dejando pasar? Imagina que este sistema, sobre las personas infectadas aciertas un 10%. Lo que significa que de todas las personas que están infectadas el sistema solo detecta el 10 de cada 100. ¡Potencialmente "estamos dejando pasar a 90 personas infectadas"!.<br>
            Una cosa que puede estar pasando aqui es que tu sistema es capaz de detectar personas infectadas en un estado muy avanzado, sin embargo otras no. Es por ello que aquí una métrica que tenga en cuenta los falsos negativos (predecir que está sano y no lo está) impacta más en el valor de negocio  u objetivo que se busca, que es que dentro no pasen infectados.
        </p>
        <p style="font-size: 18px; color: #002060;">
            <strong>Ejemplo 2:</strong> Supon que tienes una dataset super debalanceado, es decir, tienes un 95% de la clase 0 y un 5% de la clase uno. Sabiendo que la métrica de accuracy es la cantidad de aciertos sobre el número total de predicciones, ¿qué accuracy tendría en este sistema si en lugar de usar un modelo de machine learning usara un modelo dummy que siempre, siempre, prediga 0?<br>
            Pues un 95% de accuracy, trivial. La pregunta ahora es, ¿esto tiene sentido? lo más seguro que no, en estos casos la clase que más va a preocupar sea ese 5%, por tanto lo interesantes es usar métricas que ayuden a entender ese 5%.
        </p>
        <p style="font-size: 18px; color: #002060;">
            Vistos estos dos ejemplos, y entendida la importancia de las métricas. Los sistemas de recomendación tiene dos grandes grupos de métricas. <strong>Métricas <em>on-line</em></strong> y <strong>Métricas <em>off-line</em></strong>.
        </p>
        <h2 style="color: #f47b4c; margin-top: 100px;">Métricas <em>on-line</em>.</h2>
        <p style="font-size: 18px; color: #002060;">
            Como indica su nombre, son métricas en vivo, en la practica esto quiere decir que necesitas que el sistema esté en producción para poder comprobar estas métricas. Por ejemplo el CTR (<em>Click Through Rate</em>), el impacto económico de la solución o la propensión a que alguien compre un artículo recomendado.<br>
            Estas métricas son muy difíciles, por no decir imposible, de calcular en la fase de entrenamiento del modelo y de preparación de la infraestructura. Sin embargo, son métricas que se suelen usar para realizar tests A/B o para monitorizar el comportamiento de un modelo una vez ya esté en producción.
        </p>
        <p style="font-size: 18px; border: 4px solid #1cb890; text-align: center; padding: 8px; color: #002060; width: 50%; margin: 5% 25%;">
            Es por ello que es muy importante pensar, no solo en cómo hacer un buen modelo y qué métricas usar, si no en cómo voy a medir luego el impacto de lo que estoy haciendo.
        </p>
        <h2 style="color: #f47b4c; margin-top: 100px;">Métricas <em>off-line</em>.</h2>
        <p style="font-size: 18px; color: #002060;">
        Estas métricas si se pueden calcular en la fase de entrenamiento, son similares a las de cualquier proyecto de ML, sin embargo tiene un punto diferente tanto en cuanto se tiene que contar sobre el número de items recomendados. Las métricas siguen la siguiente estructura: NOMBRE@K donde K es el número de recomendaciones que se hace. Así pues, cuando se quiere evaluar el error sobre el rating que los usuarios les da a los items se usa <strong>RMSE@K</strong> o <strong>MAE@K</strong>. En el caso de que se quiera tener una sensibilidad sobre lo relevante de las recomendaciones, se usa <strong>Precision@K</strong>, <strong>Recall@K</strong> o <strong>F1@K</strong>. Si se quiere tener información sobre el orden en el que se recomienda lo que se recomienda se usa <strong>NDCG@K</strong> o <strong>MAP@K</strong>. Las métricas como <strong>AUC@K</strong> y <strong>LogLoss@K</strong> sirven para conocer el número optimo de items a recomendar (K) y para lidiar con datasets desbalanceados.
        <p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>RMSE@K</strong>. Error cuadrático medio sobre los ratings que los usuarios dan a un productos dado. O, sobre los rating que un usuario da a los productos. La variable $i$ puede representar los ratings de los usuarios sobre un producto, porque se quiera calcular la métrica sobre un producto; o los ratings que un usuario da a los diferentes productos, porque se quiera calcular sobre los usuarios. La variable $K$ es el perímetro sobre el que aplica la métrica. Y, tanto $P_{i}$ como $R_{i}$ representan la <strong>predicción</strong> del rating sobre el elemento y el valor <strong>real</strong> con el que se ha "rateado" ese elemento.
            </li>
        </ul>
        <p style="font-size: 22px; color: #002060; text-align: center;">
            $\text{RMSE@K} = \sqrt{\frac{\sum_{i=1}^{K}(P_{i} - R_{i})^{2}}{K}}$
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>MAE@K</strong>. Error medio absoluto sobre los ratings que los usuarios dan a un productos dado. O, sobre los rating que un usuario da a los productos. La variable $i$ puede representar los ratings de los usuarios sobre un producto, porque se quiera calcular la métrica sobre un producto; o los ratings que un usuario da a los diferentes productos, porque se quiera calcular sobre los usuarios. La variable $K$ es el perímetro sobre el que aplica la métrica. Y, tanto $P_{i}$ como $R_{i}$ representan la <strong>predicción</strong> del rating sobre el elemento y el valor <strong>real</strong> con el que se ha "rateado" ese elemento.
            </li>
        </ul>
        <p style="font-size: 22px; color: #002060; text-align: center;">
            $\text{MAE@K} = \frac{\sum_{i=1}^{K}\mid P_{i} - R_{i}\mid}{K}$
        </p>
        <p style="font-size: 18px; color: #002060;">
            Antes de seguir viendo métricas, vamos a definir algunos conceptos. Ya sabemos que $K$ es el número de recomendaciones, en este caso cuando se hable de recomendaciones relevantes nos vamos a referir a esas recomendaciones en las que el usuario ha realizado una interacción, por ejemplo ver una película. Y se habla de recomendaciones no relevantes en el caso contrario, aquellas recomendaciones para las cuales el usuario no interacciona. Sin embargo, puede haber productos relevantes que no se recomienden, es por ello que se define:
        <p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>RR</strong> o <em>Relevant Recommendations</em>: Son productos que <strong>se han recomendado</strong> y que <strong>han sido</strong> relevantes para el usuario.</li>
            <li style="margin: 20px 0;"><strong>IrR</strong> o <em>Irrelevant Recommendations</em>: Son productos que <strong>se han recomendado</strong> y que <strong>no han sido</strong> relevantes para el usuario.</li>
            <li style="margin: 20px 0;"><strong>RNR</strong> o <em>Relevant Non Recommendations</em>: Son productos que <strong>no se han recomendado</strong> y que <strong>han sido</strong> relevantes para el usuario.</li>
            <li style="margin: 20px 0;"><strong>IrNR</strong> o <em>Irrelevant Non Recommendations</em>: Son productos que <strong>no se han recomendado</strong> y que <strong>no han sido</strong> relevantes para el usuario.</li>
        </ul>
        <p style="font-size: 18px; color: #002060;">
            Estas definiciones son como los TN, FN, TP y FP de la matriz de confusión para un sistema de clasificación.
        <p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>Precision@K</strong>. La precisión de que los $K$ productos que se recomiendan sea relevante.</li>
        </ul>
        <p style="font-size: 22px; color: #002060; text-align: center;">
            $\text{Precision@K} = \frac{\text{RR}}{\text{RR} + \text{IrR}}$
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>Recall@K</strong>. Cuántos, de los productos relevantes, se encuentran dentro de los $K$ productos recomendados.</li>
        </ul>
        <p style="font-size: 22px; color: #002060; text-align: center;">
            $\text{Recall@K} = \frac{\text{RR}}{\text{RR} + \text{RNR}}$
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>F1@K</strong>. Media armónica entre el recall y la precisión.</li>
        </ul>
        <p style="font-size: 22px; color: #002060; text-align: center;">
            $\text{F1@K} = 2\frac{\text{Precision@K}\cdot\text{Recall@K}}{\text{Precision@K} + \text{Recall@K}}$
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>nDCG@K</strong>. Es la ganancia acumulada descontadas normalizada. Tiene en cuenta el orden de las recomendaciones. Es decir, si las recomendaciones más relevantes se sirven primero la métrica tendrá un mejor valor.</li>
        </ul>
        <p style="font-size: 22px; color: #002060; text-align: center;">
            $\text{nDCG@K} = \frac{\text{DCG@K}}{\text{DCG@K}_{+}} \longrightarrow \text{DCG@K}=\sum_{i=1}^{K}\frac{X_i}{\log_2(i + 1)}$
        </p>
        <p style="font-size: 18px; color: #002060;">
            Las variables $X_i$ son los valores de ranting. La variable de DCG@K es es la ganancia acumulada descontada, dado que tiene el término dividiendo del logaritmo pondera mejor las primeras recomendaciones. Y, $\text{DCG@K}_+$ es el mismo valor solo que con los $K$ elementos recomendados ordenados de mejor a peor.
        </p>
        <ul style="text-align: left; font-size: 18px; color: #002060;">
            <li style="margin: 20px 0;"><strong>MAP@K</strong>. Es la media de todos los usuarios sobre las medias de precisión de los productos relevantes para cada usuario. Parece un poco lioso, pero vamos a destapar la ecuación siguiente.</li>
        </ul>
        <p style="font-size: 22px; color: #002060; text-align: center;">
            $\text{MAP@K} = \frac{1}{M}\sum_{j=1}^{M}[\frac{1}{K}\sum_{i=1}^{K}\frac{\text{RR@i}}{i}\delta(i)]$
        </p>
    </body>
</html>

```python
recommendations = [1, 0, 0, 1, 1, 1, 0]
K = len(recommendations)
summ = [
    sum(recommendations[:idx])/idx for idx in range(1, K + 1)
]
summ_rel = [s*r for s, r in zip(summ, recommendations)]
(1/K) * sum(summ_rel)
```

<p style="font-size: 18px; color: #002060;">
    Imagina que la variable "recommendations" de la celda anterior son películas en orden de recomendación. El valor 1 significa que la película ha sido relevante, mientras que 0 indica que no. La variable "summ" es la parte de la ecuación que hace $\frac{\text{RR@i}}{i}$. La variable "summ_rel" es multiplicar por 1 cuando la película ha sido relevante y por 0 cuando no. Es decir, es la parte $\frac{\text{RR@i}}{i}\delta(i)$. Y, finalmente el sumatorio de estos elementos entre el total $K$ de elementos que hay. Siendo entonces que $j$ son los usuarios y $M$ es el total de usuarios que hay.
</p>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><strong>AUC@K</strong>. Es el área bajo la curva ROC para la $K$ recomendaciones. La importancia de esta métrica reside en lo siguiente: Supon que se ofrecen $K=3$ recomendaciones, de prendas, a un usuario. Pasado 1 mes, por ejemplo, recogemos la interacción de ese usuario. Se observa que ha añadido al carrito 20 productos de los cuales ha comprado 5. De esos 5 productos 1 pertenece a la recomendación que le hemos realizado. En este caso se tiene que $TP=RR=1$, $FP=IrR=2$, $TN=IrNR=13$, $FN=RNR=4$. Esto da los siguientes ratios $TPR=RRRate=\frac{RR}{RR + RNR}=\frac{1}{5}$ y $FPR=IrRRate=\frac{IrR}{IrR + IrNR}=\frac{2}{15}$. ¿Cómo generamos una curva ROC con esto? Pues si ahora en lugar de ofrecer $K=3$ artículos se ofrece $K=4,\cdots, \kappa$ y se van construyendo estos ratios y con ellos la curva final y, consecuentemente el valor de AUC.</li>
</ul>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><strong>LogLoss@K</strong>. Esta métrica solo se puede usar cuando se puede encontrar una probabilidad a que consuma un producto. Por ejemplo si se tiene que un usuario va a comprar un productos con una probabilidad del 80%, esta métrica tiene sentido. Sin embargo si lo que se está calculando es un rating, esta métrica no nos sirve. Cuanto más cerca de 0 sea esta métrica mejor, significa que el sistema está más determinado. O, en su defecto, que la probabilidad de clase difiere menos de su etiqueta real. La etiqueta $L_{i}$ es la etiqueta del registro (0 o 1), mientras que $P_{i}$ es la probabilidad asociada a ese registro.</li>
</ul>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{LogLoss@K} = -(L_{i} \log(P_{i}) + (1 - L_{i})\log(1 - P_{i}))$
</p>
<p style="font-size: 18px; color: #002060;">

<p style="font-size: 18px; color: #002060;">
    Como se ha dicho, estas métricas sirven para medir la efectividad del sistema, pero no detallan si se recomiendad novedades, si hay diferencia entre los productos recomendados, el grado de productos cubiertos por el sistema o si nos recomienda productos de una manera más aspiracional.
</p>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><strong><em>CatalogCoverage</em></strong>. Porcentaje de items recomendados sobre el total de items que hay.</li>
</ul>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{CatalogCoverage} = \frac{\text{Total de productos recomendados}}{\text{Total de productos}}$
</p>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><strong><em>DistributionalCoverage</em></strong>. Mide el grado de igualdad con el que se recomiendan artículos diferentes a los usuarios cuando se utiliza un determinado sistema de recomendación. La variable $p_{j}$ es las veces que un item aparece en las listas de cada persona, es decir, la probabilidad de aparición de un item. $p_j = \frac{\text{Total de personas a la que se le recomienda el item j}}{\text{Total de personas recomendadas}}$. N es el total de productos que hay.</li>
</ul>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{DistributionalCoverage} = -\sum_{j}^{N}p_{j}\log_2(p_{j})$
</p>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><strong><em>Diversity</em></strong>. Representa la variedad de las recomendaciones. Para ello vamos a presentar primero la similaridad del coseno, que es una medida de como dos vectores se parecen entre sí. Para esta medida $\overrightarrow{i}$ es el vector del producto $i$ donde sus elementos son los ratings (o interés) de los usuarios por el producto, $\overrightarrow{j}$ es lo mismo para el producto $j$. Mientras que $\mid\mid\overrightarrow{i}\mid\mid$ es el módulo (valor absoluto) del vector.</li>
</ul>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{Cosine Similarity(i, j)} = \frac{\overrightarrow{i}\cdot\overrightarrow{j}}{\mid\mid\overrightarrow{i}\mid\mid\cdot\mid\mid\overrightarrow{j}\mid\mid}$
</p>
<p style="font-size: 18px; color: #002060;">
    Este valor se encuentra entre -1 y 1 donde esos extremos indican que los productos son similares y 0 indican que los valores no se parecen en nada (si recordamos de la geometría, el coseno entre dos vectores es 0 cuando el angulo entre ellos es de $\frac{\pi}{2}$ radianes.) Si ahora se suma la similaridad del coseno de cada pareja de elementos recomendados ($i$, $j$) y se divide por la cantidad de parejas encontradas (${K}\choose{2}$ donde K es el número de recomendaciones) se tiene para ese usuario ($u$) la similaridad intra recomendaciones.
</p>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{IntraRecSim(i, j)}_u = \frac{1}{{K}\choose{2}}\sum_{i, j \forall i<j}\text{Cosine Similarity(i, j)}$
</p>
<p style="font-size: 18px; color: #002060;">
    Si ese valor ahora se calcula para todos los usuarios $M$ y se divide, se obtiene la similaridad intra recomendaciones
</p>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{IntraRecSim(i, j)} = \frac{1}{M}\sum_{u=1}^{M}\text{IntraRecSim(i, j)}_u$
</p>
<p style="font-size: 18px; color: #002060;">
    Y, finalmente, la <em>diversity</em> se mide como:
</p>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{Diversity} = 1 - \text{IntraRecSim(i, j)}$
</p>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><strong><em>Novelty</em></strong>. Mide cómo de novedoso es un producto o como de novedosas son las recomendaciones. Este valor se compone de la probabilidad de que un producto aparezca en el histórico de ventas (por ejemplo) $P_{i} = \frac{\text{Total de usuarios que interaccionado con el producto}}{\text{Total de usuarios}}$ y la probabilidad de que un producto aparezca recomendado a un usuario $P^{R}_{i} = \frac{\text{Total de usuarios a la que se le ha recomendado el producto}}{\text{Total de usuarios a los que se le han recomendado productos}}$. El sumatorio se hace sobre el total de productos recomendados $N^R$.</li>
</ul>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{Novelty} = -\sum_{i=1}^{N^{R}}P^{R}_{i}\log_{2}(P_{i})$
</p>
<ul style="text-align: left; font-size: 18px; color: #002060;">
    <li style="margin: 20px 0;"><strong><em>Serendipity</em></strong>. Mide como de novedosa es una recomendación para un usuario específico, comparándola con los productos con los que el usuarios ha interaccionado previamente. Antes de tener este valor, se tiene que calular el cómo de esperado es un producto para un usuario ($u$), o <em>expectedness(i)</em>. Esto no es más que la suma de las similaridades del coseno de todos los productos con los que ha interaccionado el usuario ($j$) y el producto $i$ en cuestión, dividido entre el total de productos interaccionados para el usuario.</li>
</ul>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{expectedness}(i)_u = \frac{1}{N}\sum_{j=1}^{N}\text{Cosine Similarity(i, j)}$
</p>
<p style="font-size: 18px; color: #002060;">
    Ahora bien, para calcular la <em>serendipity</em> para un usuario, sobre el total de recomendaciones K (es decir, para todos los productos recomendados) se calcula:
</p>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{Serendipity}_u = \frac{1}{K}\sum_{i=1}^{K}(1 - \text{expectedness}(i)_u)\delta(i)$
</p>
<p style="font-size: 18px; color: #002060;">
    Donde, como en la métrica MAP@K, la variable $\delta(i)$ es 1 si el usuarios reacciona positivamente con el producto recomendado y 0 si no. Finalmente, la <em>Serendipity</em> total es:
</p>
<p style="font-size: 22px; color: #002060; text-align: center;">
    $\text{Serendipity} = \frac{1}{M}\sum_{u=1}^{M}\text{Serendipity}_u$
</p>
<p style="font-size: 18px; color: #002060;">
    Y con esto, ya si, es todo.
</p>
